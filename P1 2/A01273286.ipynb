```python
import time
import pylint
#Linea para derterminar inicio de tiempo
tiempo_inicio = time.time()
#Lista maestra para las operaciones 
Master = []
numeros = ['1','2','3','4','5','6','7','8','9','0','-','.',',']
#Se lee el archivo línea por línea
with open('TC1.txt', 'r') as file:
    # Lee una línea a la vez
    for linea in file:
        aux = ""
        #Control para prevenir letras no deseadas o caracteres fuera del sistema numerico
        for i in linea.strip():
            if i in numeros:
                aux += i
            else:
                break
        Master.append(float(aux))

#Calculos para mean, median, mode, standard deviation, and variance.
Mean = 0
Varianza = 0
Desv_std = 0
Moda = 0

Suma_cuadrados_diferencias = 0
for i in Master:
    Mean += i
Mean=Mean/len(Master)
print('Count: {}'.format(len(Master)))
print('Mean: {}'.format(Mean))


for x in Master:
    #print(x)
    Suma_cuadrados_diferencias += (x-Mean)**2

Varianza = Suma_cuadrados_diferencias/(len(Master)-1)
Desv_std=Varianza**(1/2)
print("Varianza: {}".format(Varianza))
print("Desv Std: {}".format(Desv_std))


#Mediana, ordenamos los datos y comenzamos a comparar los 3 casos posibles
# Ordenar el conjunto de datos
Datos_ordenado = sorted(Master)

# Calcular la cantidad de elementos en el conjunto
Cantidad_elementos = len(Datos_ordenado)

# Calcular la mediana
if Cantidad_elementos % 2 == 1:
    # Si la cantidad de elementos es impar
    Mediana = Datos_ordenado[Cantidad_elementos // 2]
else:
    # Si la cantidad de elementos es par
    Indice1 = Cantidad_elementos // 2 - 1
    Indice2 = Cantidad_elementos // 2
    Mediana = (Datos_ordenado[Indice1] + Datos_ordenado[Indice2]) / 2

print("Mediana:", Mediana)


#Moda
#Construimos un diccionario maestro con todas las repeticiones para luego comparar la mas grande
Frecuencias={}
for elemento in Master:
    if elemento in Frecuencias:
        Frecuencias[elemento] += 1
    else:
        Frecuencias[elemento] = 1
max_frecuencia = 0

for elemento, frecuencia in Frecuencias.items():
    if frecuencia > max_frecuencia:
        max_frecuencia = frecuencia
        Moda = elemento

print("Moda: {}".format(Moda))

#Tiempo final
tiempo_final = time.time() -tiempo_inicio
print("Tiempo de ejecución:", tiempo_final, "segundos")
```
